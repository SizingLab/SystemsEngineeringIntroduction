```python
# Calcul the moment of inertia for the flywheel
M_screws= 1e-3 #Mass of each screw (kg)
N_screws=8 # Number of screw
M_nuts= 1e-3 #Mass of each nut (kg)
N_nuts=8 # Number of nuts
R_flywheel=0.021 # flywheel radius (m)

I_flywheel = (M_screws*N_screws + M_nuts*N_nuts)*R_flywheel**2 + 3.429e-6

print("The moment of inertia for the flywheel is " + "{:e}".format(I_flywheel) + " kg.m²" )

# Calcul the moment of inertia for the satellite (cube+motor+all components)
#Inertia of the motor
M_motor= 0.038 #Mass of the motor (kg)
R_motor=0.0384 #Radius of the motor (m)

I_motor=0.5*M_motor*R_motor**2 #Inertia of the motor (kg.m²)

#Inertia of the cube 1/6 * m * l**2

l_cube = 0.1 #length of the cubesat (m)
M_cube=0.1 #Mass of the cube without components (kg)

I_cube= 1/6 * M_cube * l_cube**2  #Inertia of the cube (kg.m²)

#Inertia of the lipo

l_lipo=0.075 #length of the lipo (m)
w_lipo=0.031 #width of the lipo (m)
M_lipo=0.0774 #Mass of the lipo (kg)

#I_lipo=1/12 * M_lipo * (l_lipo**2+w_lipo**2) #Inertia of the lipo (kg.m²)
I_lipo=M_lipo*(l_cube /4)**2
#Inertia of the arduino (arduino+shield)

l_arduino=0.068 #length of the arduino (m)
w_arduino=0.053 #width of the arduino (m)
M_arduino=0.066 #Mass of the arduino (kg), it's times 2 because we have the arduino and the shield

#I_arduino=1/12 * (M_arduino) * (l_arduino**2+w_arduino**2) #Inertia of the lipo (kg.m²)

I_arduino=M_arduino*(l_cube /2)**2

#Total Inertia of the satellite

I_sat_total=I_cube+I_motor+I_lipo+I_arduino #Inertia of the satellite (kg.m²)

print("The moment of inertia for the satellite is " + "{:e}".format(I_sat_total) + " kg.m²" )
```
